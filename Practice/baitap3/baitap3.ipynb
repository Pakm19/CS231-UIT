# Hàm đọc ảnh

```python
def read_image(dir_images, name_image):
  path_filename = os.path.join(dir_images, name_image)
  img = cv.imread(path_filename,0)
  return img
```

# Hàm tính Histogram

```python
def TinhHist(img):
   hist = cv.calcHist([img],[0],None, [256],[0,256])
   size = img.shape[0]*img.shape[1]
   hist = hist / size
   return hist
```

# Hàm read_data

```python
def load_data(dir_images):
  features=[]
  labels=[]
  for name_image in os.listdir(dir_images):
    feature = 
    img = read_image(dir_images, name_image)
    fimg = TinhHist(img).reshape(256)
    labels.append(int(name_image.split(".")[0]=="dog"))
    features.append(fimg)
  return np.array(features), np.array(labels)
```

# Phân loại ảnh

```python
from sklearn.neighbors import KNeighborsClassifier

features_train,labels_train = load_data(dir_images_train)
features_test,labels_test = load_data(dir_images_test)
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(features_train, labels_train)
acc = model.score(features_test, labels_test)
print(f"Độ chính xác của mô hình với k = 1: {acc}")
```
